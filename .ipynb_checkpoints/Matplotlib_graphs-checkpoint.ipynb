```python
#Creating Bar Charts Using MATLAB Method
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#If graphing nominal variables, better to use horizontal bar charts 
#Create the plot
plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
#gca() method used to "get current axes": combine that with the invert_yaxis() method
plt.gca().invert_yaxis()
plt.legend()
```

```python
#Creating Bar Charts Using the Object-oriented Method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.set_title("ByBer Fare")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()
ax.legend()
plt.show()
```

```python
#Creating a Scatter Plot Using the MATLAB Method
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Must add "o" with plt.plot() method to swtich from line chart to scatter plot
plt.plot(x_axis, y_axis, 'o')
plt.show()
```

```python
#Using the plt.scatter function instead
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title('PyBer Fare by Month')
plt.xlabel('Fare($)')
plt.ylabel('Date')
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
plt.show()
```

```python
#Creating a Bubble Chart Using the MATLAB Method
plt.scatter(x_axis, y_axis, s=y_axis)
#When set the size (s) equal to y_axis, this allows us to base size of the bubble on the datapoints
plt.show()
```

```python
#If we want to enlarge all the points, in general, we must iterate through the y-axis data and multiply each data point by 3 (x)
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s = y_axis_larger)
plt.show()
```

```python
#Using list comprehension instead to replace many 'for' and 'while' loops 
#new_list = [expression for item in list if conditional]
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
plt.show()
```

```python
#CREATING A SCATTER PLOT USING THE OBJECT-ORIENTED METHOD
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c='#87CEEB', s= [i * 5 for i in y_axis], label='Boston', alpha=0.8, edgecolor='black', linewidths=2)
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_xlim(0,50)
ax.legend()
plt.gca().invert_yaxis()
plt.show()
```
