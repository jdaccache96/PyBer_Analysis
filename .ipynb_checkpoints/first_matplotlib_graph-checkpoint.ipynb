```python
import matplotlib.pyplot as plt
```

```python
matplotlib.__version__
```

```python
#Automatically used as y-values if only give a single argument. Therefore, x-values will automatically be generated
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ['Jan','Feb','March','April','May','June','July','August','September','October','November','December']
```

```python
plt.plot(months, savings)
plt.xticks(rotation=45)
#Allows us to rotate either axis to create more room so not crowded or pushed together
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
#show() shows current plot
```
